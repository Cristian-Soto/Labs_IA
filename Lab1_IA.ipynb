### Creacion del Dataset

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
gasto_publicidad = np.random.normal(100, 20, 100)  # Gasto en publicidad (en dólares)
ventas = 50 + 3 * gasto_publicidad + np.random.normal(0, 30, 100)  # Ventas (en unidades)

# Visualizar los datos
plt.scatter(gasto_publicidad, ventas)
plt.xlabel('Gasto en Publicidad (USD)')
plt.ylabel('Ventas (unidades)')
plt.title('Relación entre el Gasto en Publicidad y las Ventas')
plt.show()
```

### Contrucción del modelo

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lin_reg = LinearRegression()
lin_reg.fit(gasto_publicidad.reshape(-1, 1), ventas)
```

```python
# Parámetro theta 0
lin_reg.intercept_
```

```python
# Parámetro theta 1
lin_reg.coef_
```

```python
# Predecir valores
ventas_pred = lin_reg.predict(gasto_publicidad.reshape(-1, 1))
```

```python
mse = mean_squared_error(ventas, ventas_pred)
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(gasto_publicidad, ventas, label='Datos')
plt.plot(gasto_publicidad, ventas_pred, color='red', label='Regresión lineal')
plt.xlabel('Gasto en Publicidad (USD)')
plt.ylabel('Ventas (unidades)')
plt.title('Gasto en Publicidad vs Ventas')
plt.legend()
text_x = np.min(gasto_publicidad) + 5
text_y = np.min(ventas) -50
plt.text(text_x, text_y, f'Ventas = {intercepto:.2f} + {coeficientes[0]:.2f} * Gasto en Publicidad\nMSE = {mse:.2f}', fontsize=10, ha='left', va='top', bbox=dict(facecolor='white', alpha=0.5))
plt.grid(True)
plt.tight_layout()
plt.show()
```
